```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Sample dataset (simulating non-linearity)
data = {
    "YearsExperience": np.arange(1, 11),
    "Salary": [40000, 43000, 48000, 55000, 63000, 75000, 88000, 102000, 118000, 135000]
}

df = pd.DataFrame(data)

# Define features and target variable
X = df[["YearsExperience"]]
y = df["Salary"]

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Create polynomial feature transformer (degree=4 for quadratic model)
poly = PolynomialFeatures(degree=2)

# Transform feature
X_poly = poly.fit_transform(X)


# Train Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X, y)

# Make predictions
y_pred_linear = linear_model.predict(X)

# Train Polynomial Regression model
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)

# Make predictions
y_pred_poly = poly_model.predict(X)


# Plot data points
plt.scatter(X, y, color='blue', label="Actual Data")

# Plot linear regression line
plt.plot(X, y_pred_linear, color='red', linestyle="dashed", label="Linear Regression")

# Plot polynomial regression curve
plt.plot(X, y_pred_poly, color='green', label="Polynomial Regression (Degree=2)")

# Formatting
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear vs Polynomial Regression")
plt.legend()
plt.show()
```
