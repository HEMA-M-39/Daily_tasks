```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Create a simple 2D dataset
np.random.seed(42)
X = np.vstack((
    np.random.normal(loc=[2, 2], scale=0.5, size=(50, 2)),
    np.random.normal(loc=[6, 6], scale=0.5, size=(50, 2)),
    np.random.normal(loc=[10, 2], scale=0.5, size=(50, 2))
))

# Step 2: Apply K-means clustering with different initial centroid positions
init_methods = ['random', 'k-means++']
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for i, init in enumerate(init_methods):
    kmeans = KMeans(n_clusters=3, init=init, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    
    # Step 3: Plot the clustered data points
    axes[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6, edgecolors='k')
    axes[i].scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
    axes[i].set_title(f"K-Means Clustering ({init} init)")
    axes[i].set_xlabel("Feature 1")
    axes[i].set_ylabel("Feature 2")
    axes[i].legend()

plt.tight_layout()
plt.show()

# Step 4: Discussion on variability due to random initialization
print("Different initial centroid selections can lead to varying clustering results, especially in cases where the clusters are not well separated.")
```
